```python
import numpy as np
def normal(y):
  count=0
  for i in y:
    if isinstance(i, list):
        y[count]=np.mean(i)
    count+=1 
  return y
def PLT(C1,C2,R1,R2,label):
  plt.figure(figsize=(20, 10))
  c1=normal(C1)
  c2=normal(C2)
  r1=normal(R1)
  r2=normal(R2)
  plt.subplot(121)
  plt.title("CPU USAGE")  
  plt.plot(size,c1,label='client')   
  plt.plot(size,c2,label='server')
  plt.xlabel('SIZE')   
  plt.legend()
  plt.grid()
  plt.subplot(122)
  plt.title("RAM USAGE")
  plt.plot(size,r1,label='client') 
  plt.plot(size,r2,label='server')   
  plt.ylabel('MB') 
  plt.xlabel('SIZE')
  plt.legend()
  plt.grid()
  plt.suptitle(label)

def PLTTtimes(CT1,CT2,ST1,ST2):
  plt.figure(figsize=(20, 10))
  plt.subplot(121)
  plt.title("CLIENT TIME DELAY")  
  plt.plot(size,CT1,label='DES')   
  plt.plot(size,CT2,label='AES') 
  plt.ylabel('SEC') 
  plt.xlabel('SIZE')    
  plt.legend()
  plt.grid()
  plt.subplot(122)
  plt.title("SERVER TIME DELAY")
  plt.plot(size,ST1,label='DES') 
  plt.plot(size,ST2,label='AES')  
  plt.ylabel('SEC') 
  plt.xlabel('SIZE')  
  plt.legend()
  plt.grid()

def PLTrams(RC1,RC2,RC3,RS1,RS2,RS3):
  plt.figure(figsize=(20, 10))
  rc1=normal(RC1)
  rc2=normal(RC2)
  rc3=normal(RC3)
  rs1=normal(RS1)
  rs2=normal(RS2)
  rs3=normal(RS3)
  plt.subplot(121)
  plt.title("CLIENT RAM USAGE")  
  plt.plot(size,rc1,label='no encrypt')   
  plt.plot(size,rc2,label='DES') 
  plt.plot(size,rc3,label='AES')  
  plt.ylabel('MB') 
  plt.xlabel('SIZE')
  plt.legend()
  plt.grid()
  plt.subplot(122)
  plt.title("SERVER RAM USAGE")
  plt.plot(size,rs1,label='no encrypt') 
  plt.plot(size,rs2,label='DES')
  plt.plot(size,rs3,label='AES')   
  plt.ylabel('MB') 
  plt.xlabel('SIZE') 
  plt.legend()
  plt.grid()
```
